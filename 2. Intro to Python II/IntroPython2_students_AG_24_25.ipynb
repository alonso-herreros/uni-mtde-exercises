# 1.&nbsp;What is NumPy?

[**Numerical Python**](http://www.numpy.org/) (usually known as **NumPy**) is the fundamental package for scientific computing with Python. It contains, among other things, the following features:

- A powerful $N$-dimensional array object
- Sophisticated broadcasting functions
- Tools for integrating C/C++ and Fortran code (this is beyond the scope of the course)
- Useful linear algebra, Fourier transform, and random number capabilities

Besides its obvious scientific uses, NumPy can also be used as an efficient multi-dimensional container for generic data. Arbitrary data types can be defined, allowing NumPy to seamlessly and efficiently integrate with a wide variety of databases.

Let's load the package and start with some examples.



````python
import numpy as np  #it's imported under the 'np' alias
````

NumPy is a Python library that allows you to work with data vectors and matrices (referred to as NumPy arrays) and directly apply operations on these arrays without the need to perform element-by-element operations.

NumPy arrays can be directly defined using methods such as `.arange()`, `.ones()`, `.zeros()`, or `.eye()`, as well as random number generators. Alternatively, they can be easily created from Python lists (or lists of lists) containing numerical elements by using `np.array(my_list)`.

At first glance, the `.flatten()`, `.ravel()` and `.reshape()` methods may seem similar, but this is not entirely accurate!

`.flatten()` always returns a copy of the original array, whereas `.ravel()` and `.reshape()` return a view of the original array whenever possible. This means that when using `.flatten()`, modifying any value of the flattened array does not affect the original array. Let's demonstrate this with an example


````python
#ORIGINAL vector:
# Let's create an array
array1 = np.array([(1,2,3,4),(3,1,4,2)])

# Let's print the original array
print ("Original Array1: ")
print(array1)


#RAVEL output:
print("\nOutput for RAVEL")
array2 = array1.ravel()
# Ravel only passes a view of
# original array to array 'b'
print('Array1 after .ravel()-> Array2 =', array2)
print('Let\'s change the first element of Array2 (Array2[0]=1000). Do Array1 and Array2 change?')
array2[0]=1000
print('Array2=', array2)
# Note here that value of original
# array 'a' at also a[0][0] becomes 1000
print('Array1 =\n', array1)
print('YES. They both have changed!')

#FLATTEN output:
print("\nOutput for FLATTEN")
array3 = array1.flatten()

# Flatten passes copy of
# original array to 'c'
print('Array1 after .flatten()-> Array3 =',array3)
print('Let\'s change the first element of Array3 (Array3[0]=5). Do Array1 and Array3 change?')
array3[0]=5
print('Array3=', array3)
# Note that by changing value of array3[0] there is no affect on value of original array1
print('Array1 =\n', array1)
print('Array3 has changed, but Array1 hasn\'t!')
````

**Object:** `.newaxis()`. Sometimes, we need to add a new dimension to an array, such as converting a 1D array into a 2D matrix. For this purpose, we can use the `.newaxis()` method.

````python
# Let's start with a 1D array:
array1 = np.array([1,1,1])
print('1D array:\n',array1)
print('Its shape is: ', array1.shape)

# Let's turn it into a column vector (2D array with dimension 1x3):
array2 = array1[:,np.newaxis]
print('\n2D array:\n',array2)
print('Its shape is: ', array2.shape)

# Let's turn it into a row vector (2D array with dimension 3x1):
array3 = array1[np.newaxis,:]
print('\n2D array:\n',array3)
print('Its shape is: ', array3.shape)
````

**Object:** `.squeeze()`. Sometimes, we might also need to remove empty or unused dimensions. To achieve this, we can use the `.squeeze()` method.



````python
print('- Original array:', array1, 'and its shape is:', array1.shape)
array1_1D = np.squeeze(array1)
print('1D array:',array1_1D, 'and its shape is:', array1_1D.shape)
print('Please note that, in this particular example, the dimensions do not change!')

print('\n- Original array:\n', array2, '\nand its shape is:', array2.shape)
array2_1D = np.squeeze(array2)
print('After squeezing...')
print('1D array:',array2_1D, 'and its shape is:', array2_1D.shape)

print('\n- Original array:', array3, 'and its shape is:', array3.shape)
array3_1D = np.squeeze(array3)
print('After squeezing... ')
print('1D array:',array3_1D, 'and its shape is:', array3_1D.shape)
````

> Note that the dimensions of `array2` and `array3` have changed!

## 1.5 Some operations with arrays

### 1.5.1 Element-wise operations

With NumPy, we can perform all standard numerical and matrix operations. For matrix addition and subtraction, the standard `+` and `-` operators can be used.



````python
x1 = np.arange(9).reshape((3, 3))
x2 = np.ones((3, 3))
result = x1 + x2
print('x1:\n', x1, '\n\nx2:\n', x2)
print('\nAddition of x1 and x2 using the + operator:\n', result)
````

However, NumPy provides built-in functions (or methods) that ensure errors and exceptions are properly handled.

````python
# We can add two matrices:
x1 = np.arange(9).reshape((3, 3))
x2 = np.ones((3, 3))
result = np.add(x1, x2)
print('x1:\n', x1, '\n\nx2:\n', x2)
print('\nAddition of x1 and x2 using built-in functions:\n', result)

# Or compute the difference:
result = np.subtract(x1, x2)
print('\nSubtraction of x1 and x2 using built-in functions:\n', result)
````

We can also add or subtract column or row vectors from arrays. Both basic operators and built-in functions will perform these operations. Unlike in Matlab, where this operation would raise an error, Python automatically applies it row by row or column by column as needed.

````python
# We can add or subtract row and column vectors:
row_vect = np.ones((1, 3))
col_vect = 2*np.ones((3, 1))
result = np.add(x1, row_vect)
print('x1:\n', x1, '\n\nrow_vect:\n', row_vect, '\n\ncol_vect:\n', col_vect)
print('\nAddition of the row vector:\n', result)
result = np.add(x1, col_vect)
print('\nAddition of the column vector:\n', result)
````

Another key difference with Matlab is that the `*` operator does not perform matrix multiplication; instead, it computes element-wise multiplication. Once again, NumPy provides a built-in function specifically for matrix multiplication, ensuring proper error handling.

````python
# We can perform element-wise multiplication by using the * operator:
x1 = np.arange(9).reshape((3, 3))
x2 = np.ones((3, 3)) * 2 # a 3x3 array with 2s in every element
result = x1 * x2
print('x1:\n', x1, '\n\nx2:\n', x2)
print('\nElement-wise multiplication of x1 and x2 using the * operator:\n', result)

# or by using the built-in NumPy function:
result = np.multiply(x1, x2)
print('\nElement-wise multiplication of x1 and x2 using a built-in function:\n', result)
````

We can use the `**` operator or the `.power()` method to raise the elements of a matrix to a specified power or to raise elements of one matrix to the corresponding powers in another matrix.




````python
M = np.array([[1,2,3],[4,5,6]])
print('Original matrix M')
print(M)
````

````python
print('M powered to 2 using the operator **:\n', M**2) #square each element of M
print('\nM powered to 2 using the .power() built-in function :\n', np.power(M,2)) #square each element of M
````

Element-wise division between matrices can be performed using the `/` operator or the `.divide()` method.

````python
print('M divided into 2 using the operator /:\n', M/2) #each element in M is divided into 2
print('\nM divided into 2 using the .divide() built-in function:\n', np.divide(M,2)) #each element in M is divided into 2
````

### 1.7.1 Pseudoinverse

In mathematics, and particularly in linear algebra, **a pseudoinverse** $A^+$ of a matrix $A$ is a generalization of the inverse matrix. The most widely known type of pseudoinverse is the [Mooreâ€“Penrose inverse](http://mathworld.wolfram.com/Moore-PenroseMatrixInverse.html).

When $A$ has linearly independent rows, i.e., when $AA^*$ is invertible where $A^*$ is the Hermitian transpose, then the pseudoinverse $A^+$ can be computed as follows:

$$ A^+ = A^* (AA^*)^{-1} $$

and it verifies that $AA^+ =I$.

Let's now compute the pseudoinverse of `M`.

````python
M2PSI = np.linalg.pinv(M)
print('The pseudoinverse of M is:\n ', M2PSI)
````

Letâ€™s verify that the result is correct!

````python
print('Sum of the diagonal elements:', round(np.trace(M @ M2PSI),1))
print('Sum of the out diagonal elements:',round(np.trace(M @ M2PSI)-np.sum(M @ M2PSI),1)) #Â Sum of out diagonal elements
````

## 3.1  Data generation

In Python, random samples can be easily generated using the `numpy.random` package. Within it, you'll find many useful tools for sampling from the most important probability distributions.

We have several common number generator functions, such as:
* `numpy.random.rand(d0,d1,...,dn)`: It **uniformly** generates random samples.
* `numpy.random.randn(d0,d1,...,dn)`: It generates samples from a **standard normal** distribution.

Please note that `d0,d1,...,dn` represent the dimensions of the returned array.

Or more specific ones:
* `numpy.random.exponential([scale, size])`: It draws samples from an exponential distribution with a specified scale parameter.

  Please note that the scale parameter in the `numpy.random.exponential()` function corresponds to the inverse of the rate ($\lambda$) in the exponential distribution:

$$p(x|\beta) = \frac{1}{\beta} \exp \Big(-\frac{x}{\beta}\Big)u(x)$$



* `numpy.random.normal([loc, scale, size])`:	It generates random samples from a normal (Gaussian) distribution with parameters: `loc` (mean of the distribution) and `scale` (standard deviation of the distribution).
* `numpy.random.uniform([low, high, size])`: Samples are uniformly distributed over the half-open interval $[$ `low`, `high`$)$ (includes `low`, but excludes `high`). In other words, any value within the given interval is equally likely to be drawn by uniform.

Please note that `size` denotes the output shape (a `int` or `tuple of ints`). If the given shape is, e.g., $(m, n, k)$, then $m * n * k$ samples are drawn.




In the following examples, we will explore different random generation methods and visualize the results. For now, you can ignore the visualization code; we will cover how these visualization tools work later.

````python
# Random sampling examples
n = 1000 # number of samples

# Sampling from a standard uniform distribution:
x_unif = np.random.rand(n)

fig1 = plt.figure()
plt.hist(x_unif, bins=100)
plt.title('Samples from a uniform distribution between 0 and 1')
plt.grid()
plt.show()

# Sampling from a normal distribution:
x_norm = np.random.randn(n)

fig2 = plt.figure()
plt.hist(x_norm, bins=100)
plt.title('Samples from a normal distribution with 0 mean and unity variance')
plt.grid()
plt.show()

# Adding Gaussian noise to a linear function:
n = 30
x = np.linspace(-5, 5, n)
noise = np.random.randn(n)
y = 3*x
y_noise = y + noise

fig3 = plt.figure()
plt.plot(x, y, color='black', linestyle='--', label='Clean signal')
plt.plot(x, y_noise, color='red', label='Noisy signal')
plt.legend(loc=4, fontsize='large')
plt.title('Visualization of a noisy data-set')
plt.grid()
plt.show()

#See https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html for more options!
````

More about plotting with `matplotlib`.

Consider `x` as a vector containing samples from a 1D random variable. A histogram is a plot that represents the observed frequencies for different ranges of the `x` domain. These frequencies can be expressed as relative frequencies (summing to $1$) or absolute frequencies (counting the number of events).

We can adjust the number and size of the intervals (called bins), which directly affects the resolution of the plot.

When we have a sufficiently large number of random samples from the same distribution, the histogram is expected to resemble the theoretical shape of the probability density function for that distribution.

In Matplotlib, we have already plotted histograms using `plt.hist(samples,bins)`.

Let's take a look at some examples.

````python
# We take samples from a normalized gaussian distribution, and we change
# mean and variance with an operation.
sigma = 4
mn = 5
#x_gauss = mn + np.sqrt(sigma)*np.random.randn(5000)
x_gauss = np.random.normal(mn, sigma/2, 5000) #(easy way!!!)


# Let's obtain an histogram with high resultion, that is, a lot of bins.
fig1 = plt.figure()
plt.hist(x_gauss, bins=100,label='Samples')
plt.title('Histogram with 100 bins')

# With vertical lines, we plot the mean and the intervals obtain summing one
# standard deviation to the mean.
plt.axvline(x=np.mean(x_gauss),color='k',linestyle='--',label='Mean')
plt.axvline(x=np.mean(x_gauss)+np.std(x_gauss),color='grey',linestyle='--',label='Mean +/- std')
plt.axvline(x=np.mean(x_gauss)-np.std(x_gauss),color='grey',linestyle='--')


plt.legend(loc='best')
plt.show()

# We check that the mean and variance of the samples is aprox. the original one.
print('Sample mean = ', x_gauss.mean())
print('Sample variance = ', x_gauss.var())

# Now let's plot a low resolution histogram, with just a few bins.
fig2 = plt.figure()

# We can set weights in this way to obtain a normalized histogram, i.e.,
# expressing relative frequencies.
w_n = np.zeros_like(x_gauss) + 1. / x_gauss.size

plt.hist(x_gauss, bins=10,label='Samples',weights=w_n)
plt.title('Histogram with 10 bins')
plt.axvline(x=np.mean(x_gauss),color='k',linestyle='--',label='Mean')
plt.axvline(x=np.mean(x_gauss)+np.std(x_gauss),color='grey',linestyle='--',label='Mean +/- std')
plt.axvline(x=np.mean(x_gauss)-np.std(x_gauss),color='grey',linestyle='--')
plt.legend(loc='best')
plt.show()

# A different resolution leads to different representations, but don't forget
# that we are plotting the same samples.

print('Sample mean = ', x_gauss.mean())
print('Sample variance = ', x_gauss.var())
````

### 3.2.2 Array of plots and text over figures

You can easily create arrays of plots to visualize them together. Here's an example. We also demonstrate how to include annotations in the figures.

````python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
num_figure = 1
plt.annotate('local max', xy=(1, f(1)), xytext=(1.25, 0.85),
             arrowprops=dict(facecolor='red', width = 1))
plt.text(4, 0.75, "Figura {0:d}".format(num_figure))
plt.grid()

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.text(6, 0, "Figure {0:d}".format(num_figure+1), style='italic',
        bbox={'facecolor': 'blue', 'alpha': 0.1, 'pad': 10})
plt.grid()
plt.show()
````

### 3.2.3 Contour and surface plots

These plots are useful for representing functions over 2D spaces.


````python
#Â We consider the following function with 2D inputs

def f(x,y):
    return np.exp((+x**2-y**2+x*y+x-y)/20)
````

````python
from mpl_toolkits.mplot3d import Axes3D

# Lets start with a contour plot

grid_points = 100
lims = 4
# Grid of grid_points**2 points in the [-2,2] square
x,y = np.mgrid[-lims:lims:2*lims/grid_points, -lims:lims:2*lims/grid_points]
grid = np.transpose(np.vstack([x.ravel(), y.ravel()]))

# Lets evaluate the function over the grid

f_grid = f(grid[:,0],grid[:,1]).reshape([grid_points,grid_points])

fig = plt.figure(figsize=(6, 18))

# Contour lines
plt.subplot(311)
plt.contour(x,y,f_grid,np.arange(0,1,0.05),cmap='coolwarm') # We plot the 0, 0.1, 0.2, ..., 1.0 contour lines
plt.colorbar()
plt.grid()
plt.title("Contour lines")

# Contour plot
plt.subplot(312)
plt.contourf(x,y,f_grid,cmap='coolwarm')
plt.colorbar()
plt.grid()
plt.title("Contour plot")

# Surface plot
ax = fig.add_subplot(3, 1, 3, projection='3d')
ax.plot_surface(x,y,f_grid,cmap='coolwarm')
plt.colorbar()
plt.grid()
plt.title("Surface plot")

plt.show()
````
